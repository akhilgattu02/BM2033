```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## When X1 and X2 are independent identical distributions

```python
#Covariance matrix
matrix = [[1, 0], [0, 10]]

mean = 0

#Y is column vector
Y = lambda x,y: np.transpose([[x-mean, y-mean]])

#Inverse of covariance matrix
Inv = np.linalg.inv(matrix)
f_z = []


x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
x_1 = []
y_1 = []
for d in x:
    for e in y:
        x_1.append(d)
        y_1.append(e)
        t = np.dot(np.transpose(Y(d,e)), Inv)
        a = np.dot(t, Y(d,e))
        gauss = ((np.e)**(-a[0][0]/2))/(2*np.pi)
        f_z.append(gauss)
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.scatter(x_1, y_1, f_z, color='black')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Gaussian when X1 and X2 are i.i.d');
```

```python
#Covariance matrix
matrix = [[10, 0.3], [0.3, 1]]

mean = 0

#Y is column vector
Y = lambda x,y: np.transpose([[x-mean, y-mean]])

#Inverse of covariance matrix
Inv = np.linalg.inv(matrix)
f_z = []


x = np.linspace(-4, 4, 20)
y = np.linspace(-4, 4, 20)
x_1 = []
y_1 = []
for d in x:
    for e in y:
        x_1.append(d)
        y_1.append(e)
        t = np.dot(np.transpose(Y(d,e)), Inv)
        a = np.dot(t, Y(d,e))
        gauss = ((np.e)**(-a[0][0]/2))/(2*np.pi*(np.sqrt(0.91)))
        f_z.append(gauss)
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
ax.scatter(x_1, y_1, f_z, color='black')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Gaussian when X1 and X2 are positively correlated');
```
